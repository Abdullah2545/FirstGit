```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Chilla_data2_for_plots.csv')
```

```python
data.head()
```

## Line plot with Title

```python
sns.lineplot(x="Age",y="Your Weight in kg? (float)",data = data)
plt.title('Line plot (Weight vs Age)')
plt.show()
```

# Bar plot

```python
kashti = pd.read_csv('titanic.csv')
kashti.head()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="sex",y="survived", hue='pclass',data=kashti)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="sex",y="survived", hue='pclass',data=kashti,order=['female','male'])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="sex",y="survived", hue='pclass',data=kashti,order=['female','male'],color='brown',ci=None)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="sex",y="fare", hue='pclass',data=kashti,order=['female','male'],estimator=np.median,color='brown',ci=None,palette='pastel')
plt.show()
```

```python

plt.figure(figsize=(8,6))
sns.barplot(x="sex",y="survived", hue='pclass',data=kashti,order=['female','male'],color='brown',estimator=np.mean,palette='pastel')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="sex",y="fare", hue='pclass',data=kashti,order=['female','male'],color='brown',estimator=np.median,palette='pastel')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="sex",y="fare", hue='pclass',data=kashti,order=['female','male'],color='brown',
            estimator=np.median,palette='pastel',saturation=0.3)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="fare",y="sex", hue='pclass',data=kashti,order=['female','male'],color='brown',
            estimator=np.median,palette='pastel',saturation=0.8)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="sex",y="fare", hue='pclass',data=kashti,
           linewidth=3.5,facecolor=(1,0.2,1,1),
           errcolor=".6",edgecolor='.4')
plt.show()
```

## custom bar plots


```python
sns.set(style='whitegrid')
plt.figure(figsize=(8,6))
sns.boxplot(x="sex",y="age",data=kashti,
            showmeans=True,
            meanprops={'marker':'*',
                      'markersize':'12',
                      'markeredgecolor':'yellow'},
           linewidth=3.5,)
plt.title('Titanic',size=26,weight='bold')
plt.xlabel('Sex (Gender)',size=16,weight='bold')
plt.ylabel('Age',size=16,weight='bold')
plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(8,6))
g = sns.FacetGrid(kashti,col = 'pclass')
g.map(sns.boxplot,x=kashti["sex"],y=kashti["age"],showmeans=True,
          meanprops={'marker':'*',
                     'markersize':'12',
                     'markeredgecolor':'yellow'})
# sns.boxplot(x="sex",y="age",data=kashti,
#             showmeans=True,
#             meanprops={'marker':'*',
#                       'markersize':'12',
#                       'markeredgecolor':'yellow'},
#            linewidth=3.5,)
# plt.title('Titanic',size=26,weight='bold')
# plt.xlabel('Sex (Gender)',size=16,weight='bold')
# plt.ylabel('Age',size=16,weight='bold')
plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(8,6))
g = sns.FacetGrid(kashti,col = 'pclass')
g.map_dataframe(sns.boxplot,"sex","age",showmeans=True,
          meanprops={'marker':'*',
                     'markersize':'12',
                     'markeredgecolor':'yellow'})
g.set_axis_labels("Sex(Gender)",'Age')
g.set_titles(col_template='{col_name} Fare')
# sns.boxplot(x="sex",y="age",data=kashti,
#             showmeans=True,
#             meanprops={'marker':'*',
#                       'markersize':'12',
#                       'markeredgecolor':'yellow'},
#            linewidth=3.5,)
# plt.title('Titanic',size=26,weight='bold')
# plt.xlabel('Sex (Gender)',size=16,weight='bold')
# plt.ylabel('Age',size=16,weight='bold')
plt.show()
```
